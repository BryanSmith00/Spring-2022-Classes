# DBSCAN
---

DBSCAN has several advantages over K-Mean clustering, primarily being it works on more datasets and more irregular  patterns and provides more robust results. We also don't need to provide it with a number of clusters which cuts out a major step of K-means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5
sns.set_style('darkgrid')
```

```python
customer_df = pd.read_csv('customers.csv')
customer_df.drop(['CustomerID'], axis=1, inplace=True)
customer_df.head()
```

```python
customer_df.info()
```

All cells are non null and we only have 1 categorical variable

```python
sns.heatmap(customer_df.corr(), annot=True)
plt.show()
```

No strong collinearity to deal with

```python
plt.pie(customer_df['Genre'].value_counts(), labels=['Female','Male'], colors=['Pink','Blue'], shadow=True,
        explode=[0,0.1])
plt.legend()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Slightly more female than male entries but should not be significant enough a difference to skew the data

```python
sns.pairplot(customer_df, hue='Spending_Score', vars=['Age', 'Annual_Income_(k$)'])
plt.show()
```

```python
# Saving only Spending_Score and income values into X.
X = customer_df.iloc[:, [2, 3]].values
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=3, min_samples=4, metric='euclidean')
db.fit(X)
```

Creates the model with parameters eps - the maximum distance between two samples for a neighborhood classification, min_samples - the number of samples in a neighborhood for a point to be considered core, and metric - method used to calculate distance (Euclidean is default like we discussed in class)

```python
db.labels_
```

```python
y_preds = db.fit_predict(X)

plt.scatter(X[y_preds == 0, 0], X[y_preds == 0, 1], s = 50, c = 'pink', label='cluster 1')
plt.scatter(X[y_preds == 1, 0], X[y_preds == 1, 1], s = 50, c = 'yellow', label='cluster 2')
plt.scatter(X[y_preds == 2, 0], X[y_preds == 2, 1], s = 50, c = 'cyan', label='cluster 3')
plt.scatter(X[y_preds == 3, 0], X[y_preds == 3, 1], s = 50, c = 'magenta', label='cluster 4')
plt.scatter(X[y_preds == 4, 0], X[y_preds == 4, 1], s = 50, c = 'orange', label='cluster 5')
plt.scatter(X[y_preds == 5, 0], X[y_preds == 5, 1], s = 50, c = 'blue', label='cluster 6')
plt.scatter(X[y_preds == 6, 0], X[y_preds == 6, 1], s = 50, c = 'red', label='cluster 7')
plt.scatter(X[y_preds == 7, 0], X[y_preds == 7, 1], s = 50, c = 'black', label='cluster 8')
plt.scatter(X[y_preds == 8, 0], X[y_preds == 8, 1], s = 50, c = 'violet', label='cluster 9')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

From this plot we can see the 9 neighborhoods DBSCAN separated our data into. A nice thing about this method is the automatic generation of the number of clusters based on our parameters and the data.
